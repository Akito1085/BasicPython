## 1問目
使うcsvはhist_data.csvです。下にコードを書いてください。

![問1](./imgs/matplotlib/plot_1.png)

```python
hist_data=pd.read_csv("data/hist_data.csv")

#データの描画
plt.hist(hist_data["aデータ"] ,label="aデータ",bins=30,alpha=0.5,color="orange")
plt.hist(hist_data["bデータ"] ,label="bデータ",bins=30,alpha=0.3,color="cyan")

#データのラベル
plt.title("aデータとbデータの温度に関する度数分布")
plt.xlabel("温度(℃)")
plt.ylabel("度数")
plt.legend()

```

## 2問目
使うcsvはscatter_data.csvです。下にコードを書いてください。

![問2](./imgs/matplotlib/plot_2.png)

```python
#データの読み込み
scatter_data=pd.read_csv("data/scatter_data.csv")
#classの定義
class0=scatter_data[scatter_data["class"]==0]
class1=scatter_data[scatter_data["class"]==1]
class2=scatter_data[scatter_data["class"]==2]

#labelの表示
plt.title("クラス0～2のlengthとwidthに関する散布図")
plt.xlabel("length(cm)", fontsize=11)
plt.ylabel("width(cm)", fontsize=11)

#範囲を指定
plt.xlim(4.1,8.1)
plt.ylim(1.9,4.52)
plt.scatter(class0['length (cm)'], class0['width (cm)'], label='class0', color='r')
plt.scatter(class1['length (cm)'], class1['width (cm)'], label='class1', color='b')
plt.scatter(class2['length (cm)'], class2['width (cm)'], label='class2', color='g')

#散布図の表示
plt.legend()
plt.grid(True)
plt.show()
```

## 3問目
使うcsvはline_data.csvです。下に2つの画像を表示するコードを書いてください。

![問3_1](./imgs/matplotlib/plot_3_1.png)

![問3_2](./imgs/matplotlib/plot_3_2.png)

```python
line_data=pd.read_csv("data/line_data.csv")
#グラフ1
#Label
plt.title("期間ごとの件数の推移")
plt.xlabel("期間")
plt.ylabel("件数")
plt.xticks(rotation=30)

#grid線
plt.grid(True)
#グラフを描画
plt.plot(line_data["期間"],line_data["件数"],marker="o")
plt.show()

#グラフ2
#label
plt.title("期間ごとの費用の推移")
plt.xlabel("期間")
plt.ylabel("費用(円)")
plt.xticks(rotation=30)

#grid線
plt.grid(True)

#グラフを描画
plt.bar(line_data["期間"],line_data["費用"])
plt.show()
```
